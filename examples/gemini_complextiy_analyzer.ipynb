```python

def test_complexity_analyzer(analyzer, test_cases):
    """
    Tests the complexity analyzer by comparing actual results with expected complexity scores and models.
    
    :param analyzer: Instance of the complexity analyzer.
    :param test_cases: List of test cases containing prompt, expected complexity, and expected model.
    """
    total_tests = len(test_cases)
    passed_tests = 0

    for i, test in enumerate(test_cases, start=1):
        prompt = test["prompt"]
        expected_complexity = test["expected_complexity"]
        expected_model = test["expected_model"]

        # Run the complexity analyzer on the prompt
        result = analyzer.get_model_recommendation(prompt)

        actual_complexity = result.get("overall_complexity", -1)
        actual_model = result.get("recommended_model", "Unknown")

        # Check complexity within a tolerance range (Â±5)
        complexity_match = abs(actual_complexity - expected_complexity) <= 5
        model_match = actual_model == expected_model

        # Log mismatches
        if not model_match and actual_complexity < expected_complexity:
            print(f"Test {i}/{total_tests} FAILED âŒ")
            print(f"Prompt: {prompt[:100]}...")  # Show only first 100 chars for readability
            print(f"Expected Complexity: {expected_complexity}, Actual: {actual_complexity}")
            print(f"Expected Model: {expected_model}, Actual: {actual_model}\n")
            print()
        else:
            print(f"Test {i}/{total_tests} PASSED âœ…")

        # Count passed tests
        if complexity_match and model_match:
            passed_tests += 1

    # Summary
    print(f"Test Summary: {passed_tests}/{total_tests} tests passed.")
```
